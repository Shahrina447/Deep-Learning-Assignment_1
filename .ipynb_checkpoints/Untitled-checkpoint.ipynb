```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


# COMMON 1 — LOAD DATASET

def load_dataset(path):
    data = pd.read_csv(path)
    X = data.iloc[:, :-1].values
    y = data.iloc[:, -1].values
    return X, y


# COMMON 2 — NORMALIZATION

def normalize(trainX, testX):
    scaler = StandardScaler()
    trainX = scaler.fit_transform(trainX)
    testX = scaler.transform(testX)
    return trainX, testX


# COMMON 3 — ADD BIAS COLUMN

def add_bias(X):
    return np.c_[np.ones((X.shape[0], 1)), X]


# COMMON 4 — PLOTTING COST CURVES

def plot_costs(train, test, title):
    plt.plot(train, label="Train Cost")
    plt.plot(test, label="Test Cost")
    plt.title(title)
    plt.xlabel("Epoch")
    plt.ylabel("Cost")
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python

# LINEAR 1 — HYPOTHESIS

def h_linear(X, theta):
    return X @ theta


# LINEAR 2 — COST (MSE)

def mse_cost(X, y, theta):
    m = len(y)
    return np.sum((h_linear(X, theta) - y) ** 2) / (2 * m)

# LINEAR 3 — GRADIENT DESCENT

def gradient_descent_linear(X_train, y_train, X_test, y_test, lr, epochs):
    m, n = X_train.shape
    theta = np.zeros(n)

    train_curve = []
    test_curve = []

    for _ in range(epochs):
        error = h_linear(X_train, theta) - y_train
        gradient = (1/m) * (X_train.T @ error)
        theta -= lr * gradient

        train_curve.append(mse_cost(X_train, y_train, theta))
        test_curve.append(mse_cost(X_test, y_test, theta))

    return theta, train_curve, test_curve
```

```python

# LOGISTIC 1 — SIGMOID

def sigmoid(z):
    return 1 / (1 + np.exp(-z))


# LOGISTIC 2 — HYPOTHESIS

def h_logistic(X, theta):
    return sigmoid(X @ theta)


# LOGISTIC 3 — COST (Cross Entropy)

def ce_cost(X, y, theta):
    m = len(y)
    pred = h_logistic(X, theta)
    pred = np.clip(pred, 1e-10, 1 - 1e-10)
    return -np.mean(y*np.log(pred) + (1-y)*np.log(1-pred))


# LOGISTIC 4 — GRADIENT DESCENT

def gradient_descent_logistic(X_train, y_train, X_test, y_test, lr, epochs):
    m, n = X_train.shape
    theta = np.zeros(n)

    train_curve = []
    test_curve = []

    for _ in range(epochs):
        preds = h_logistic(X_train, theta)
        gradient = (1/m) * (X_train.T @ (preds - y_train))
        theta -= lr * gradient

        train_curve.append(ce_cost(X_train, y_train, theta))
        test_curve.append(ce_cost(X_test, y_test, theta))

    return theta, train_curve, test_curve
```

```python
path = "BostonHousing.csv"

# Load Data
X, y = load_dataset(path)

# Split
X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)

# Normalize + Bias
X_train, X_test = normalize(X_train_raw, X_test_raw)
X_train = add_bias(X_train)
X_test = add_bias(X_test)

# Train Linear Model
theta, tr, ts = gradient_descent_linear(X_train, y_train, X_test, y_test,
                                        lr=0.01, epochs=500)

plot_costs(tr, ts, "Linear Regression Cost Curve")
```

```python
# Convert to Binary
median_val = np.median(y)
y_bin = (y > median_val).astype(int)

# Split
X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X, y_bin, test_size=0.3, random_state=42)

# Normalize + Bias
X_train, X_test = normalize(X_train_raw, X_test_raw)
X_train = add_bias(X_train)
X_test = add_bias(X_test)

# Train Logistic Model
theta, tr, ts = gradient_descent_logistic(X_train, y_train, X_test, y_test,
                                          lr=0.1, epochs=2000)

plot_costs(tr, ts, "Logistic Regression Cost Curve")
```
